## Web Scraping: BeautifulSoup

Scrabing the data from Aqar website

##### Imoprting Libraries & Methods

```python
from bs4 import BeautifulSoup
import requests
import time,os
import pandas as pd
import numpy as np
import re
```

```python
def get_datalist(findall_soup):
    return [f.text for f in findall_soup]
```

##### Inputing the URL :

```python
url = "https://sa.aqar.fm/%D9%81%D9%84%D9%84-%D9%84%D9%84%D8%A8%D9%8A%D8%B9/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6"
response = requests.get(url)
response.status_code #200 = success!
```

#### Scraping Multiple Pages

```python
site_dict = {}
df_list = []
search_list = [
    "https://sa.aqar.fm/%D9%81%D9%84%D9%84-%D9%84%D9%84%D8%A8%D9%8A%D8%B9/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6",

   
]
# for loop to take the web pages from 1 to 75 
for index, u in enumerate(search_list):
    for i in range(1,76):
        if i == 1:
            page_num = ""
        else:
            page_num = str(i)

        url = u + "/" + page_num
        response = requests.get(url)
        print(response.status_code)
        if response.status_code != 200:
            continue
        
        #Request HTML and parse
        page = response.text
        soup = BeautifulSoup(page, 'lxml')
        
        loc_soup = soup.find(class_='main-content col-centered main-content-width').find_all('span', class_='locText')
        price_soup = soup.find(class_='main-content col-centered main-content-width').find_all('span', class_='price')
        size_soup =soup.find(class_='main-content col-centered main-content-width').find_all('span', class_='size')
        bath_soup  = soup.find(class_='main-content col-centered main-content-width').find_all('span', class_='bath')
        bed_soup= soup.find(class_='main-content col-centered main-content-width').find_all('span', class_='bed')
        
        # make all colom in dec
        site_dict["location"] = get_datalist(loc_soup)
        site_dict["price"] = get_datalist(price_soup)
        site_dict["size"] = get_datalist(size_soup)
        site_dict["bath"] = get_datalist(bath_soup)
        site_dict["bed"] = get_datalist(bed_soup)
        site_dict["search"] = [index for ii in loc_soup]
        site_dict["page"] = [i for ii in loc_soup]
        site_dict["link"] = [u for ii in loc_soup]


        df_local = pd.DataFrame(site_dict)
        df_list.append(df_local)
        
         #Wait for page to load
        time.sleep(3)
    
df = pd.concat(df_list)
```
